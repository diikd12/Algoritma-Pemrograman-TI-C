# Bubble Sort - Ascending

```python
def bubble_sort_asc(data):
  n = len(data)

  for i in range(0, n - 1):
    for j in range(n - 1, i, -1):
      if data[j] < data[j - 1]:
        data[j], data[j - 1] = data[j - 1], data[j]

  return data
```

```python
data = [6, 3, 9, 1, 5]

print("Data sebelum disort:", data)
bubble_sort_asc(data)
print("Data setelah disort:", data)
```

# Bubble Sort - Descending

```python
def bubble_sort_desc(data):
  n = len(data)

  for i in range(0, n - 1):
    for j in range(0, n - i - 1):
      if (data[j] < data[j + 1]):
        data[j], data[j + 1] = data[j + 1], data[j]

  return data
```

```python
data = [6, 3, 9, 1, 5]

print("Data sebelum disort:", data)
bubble_sort_asc(data)
print("Data setelah disort:", data)
```

# Selection Sort - Maximum

```python
def max_selection_sort(data):
  n = len(data)

  for i in range(0, n - 1):
    max_index = 0

    for j in range(1, n - i):
      if data[j] > data[max_index]:
        max_index = j

    data[max_index], data[n - 1 - i] = data[n - 1 - i], data[max_index]

  return data
```

```python
data = [6, 3, 9, 1, 5]

print("Data sebelum disort:", data)
bubble_sort_asc(data)
print("Data setelah disort:", data)
```

# Selection Sort - Minimum

```python
def min_selection_sort(data):
  n = len(data)

  for i in range(0, n - 1):
    min_index = i

    for j in range(i+1, n):
      if data[j] < data[min_index]:
        min_index = j

    data[min_index], data[n - 1 - i] = data[n - 1 - i], data[min_index]

  return data
```

```python
data = [6, 3, 9, 1, 5]

print("Data sebelum disort:", data)
bubble_sort_asc(data)
print("Data setelah disort:", data)
```
